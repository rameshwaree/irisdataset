```python
#importing the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering, MeanShift
from sklearn.mixture import GaussianMixture

# Loading the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names

# Converting the DataFrame for easier handling
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target
```

```python
#plotting variety clusters
plt.figure(figsize=(12, 6))
plt.scatter(data[:, 2], data[:, 3], c=target, cmap='viridis', edgecolor='k', s=150)
plt.title('Actual Iris Varieties')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

```python
#hac clustering
hac = AgglomerativeClustering(n_clusters=3)
hac_clusters = hac.fit_predict(data)

plt.figure(figsize=(12, 6))
plt.scatter(data[:, 2], data[:, 3], c=hac_clusters, cmap='viridis', edgecolor='k', s=150)
plt.title('Hierarchical Agglomerative Clustering (HAC)')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

```python
#meanshift clustering
mean_shift = MeanShift()
mean_shift_clusters = mean_shift.fit_predict(data)

plt.figure(figsize=(12, 6))
plt.scatter(data[:, 2], data[:, 3], c=mean_shift_clusters, cmap='viridis', edgecolor='k', s=150)
plt.title('MeanShift Clustering')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

```python
#gmm clustering
gmm = GaussianMixture(n_components=3)
gmm_clusters = gmm.fit_predict(data)

plt.figure(figsize=(12, 6))
plt.scatter(data[:, 2], data[:, 3], c=gmm_clusters, cmap='viridis', edgecolor='k', s=150)
plt.title('Gaussian Mixture Model Clustering')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```
